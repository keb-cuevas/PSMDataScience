#Looped Code

```python
for url in reversed(urls_list2):
  pdf = tb.read_pdf(url, pages='all', stream=True)
  df = pdf[0].reset_index().T.reset_index().T

  new_header = df.iloc[0] #grab the first row for the header
  df = df[1:] #take the data less the header row
  df.columns = new_header #set the header row as the df header

  cols = df.columns.to_list()[1], df.columns.to_list()[-4], df.columns.to_list()[-3], df.columns.to_list()[-2], df.columns.to_list()[-1] #last 4 columns

  #cols = df.columns.to_list()[1], df.columns.to_list()[-2], df.columns.to_list()[-1] #last 2 columns
  df = df[df.columns[df.columns.isin(cols)]].reset_index(drop=True)

  df = df[1:]

  s = url
  result = re.search('https://www.da.gov.ph/wp-content/uploads/2018/03/(.*).pdf', s)
  date1 = result.group(1)
  date = date1.split('-')[-3:]
  date2 = date
  df[['Month', 'Day', 'Year']] = date2

  df.columns=['Commodity', 'Prevailing', 'Low', 'High', 'Average', 'Month', 'Day', 'Year']
  df = df.dropna()

  df = df[['Month', 'Day', 'Year', 'Commodity', 'Low', 'High', 'Average', 'Prevailing']]

  folder_location = r'/content/drive/MyDrive/Prices'
  if not os.path.exists(folder_location):os.mkdir(folder_location)

  df.to_csv(f'/content/drive/MyDrive/Prices/{result.group(1)}.csv', sep=',', index=False)
```
